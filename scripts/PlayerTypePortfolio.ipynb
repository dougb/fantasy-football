```python
import pandas as pd
import seaborn as sns
import random
import numpy as np

%matplotlib inline
```

Make a dataframe for stats of players for a particular position that played in 2014.

```python
def make_df(position):
    df = pd.read_csv("/home/juliet/src/football/data/superNovaOut/" + position + ".txt", header=None, sep="\t",
                    names=["pid", "season", "totalPts", "stdDev"])
    df = df[df.season==2014][["pid", "totalPts", "stdDev"]].drop_duplicates().sort(columns="totalPts", ascending=False)
    return df[df.totalPts >0]
```

Define a plot that takes a dataframe with columns 'totalPts' and 'stdDev' and create a plot of the individual and joint distributions using kenrel density estimation.

```python
def joint_kde_plot(df):
    with sns.axes_style("white"):
            plot = sns.jointplot(df.totalPts, df.stdDev, kind="kde")
```

```python
te_df = make_df("TE")
joint_kde_plot(te_df)
```

```python
rb_df = make_df("RB")
joint_kde_plot(rb_df)
```

```python
qb_df = make_df("QB")
joint_kde_plot(qb_df)
```

```python
wr_df = make_df("WR")
joint_kde_plot(wr_df)
```

```python
def some(x, n):
    return x.ix[random.sample(x.index, n)]
def joinable_sample(df):
    sample = some(df, 5)
    sample['dummy'] = 0
    return sample
```

```python
wr1 = joinable_sample(wr_df)
wr1.columns = ['wr1', 'wr1_total_pts', 'wr1_std_dev', 'dummy']
wr2 = joinable_sample(wr_df)
wr2.columns = ['wr2', 'wr2_total_pts', 'wr2_std_dev', 'dummy']
rb = joinable_sample(rb_df)
rb.columns = ['rb', 'rb_total_pts', 'rb_std_dev', 'dummy']
te = joinable_sample(te_df)
te.columns = ['te', 'te_total_pts', 'te_std_dev', 'dummy']
qb = joinable_sample(qb_df)
qb.columns = ['qb', 'qb_total_pts', 'qb_std_dev', 'dummy']
```

```python
merge1 = pd.merge(wr1, wr2, on='dummy')
merge2 = pd.merge(merge1, rb, on='dummy')
merge3 = pd.merge(merge2, te, on='dummy')
all_pos = pd.merge(merge3, qb, on='dummy')
del all_pos['dummy']
```

```python
team_stats = pd.DataFrame()
team_stats['wr1'] = all_pos['wr1']
team_stats['wr2'] = all_pos['wr2']
team_stats['rb'] = all_pos['rb']
team_stats['te'] = all_pos['te']
team_stats['qb'] = all_pos['qb']
team_stats['x'] = all_pos['wr1_total_pts']+ all_pos['wr2_total_pts'] + all_pos['rb_total_pts'] + all_pos['qb_total_pts'] + all_pos['te_total_pts']
team_stats['y'] = np.sqrt(all_pos['wr1_std_dev']**2 + all_pos['wr2_std_dev']**2 + all_pos['rb_std_dev']**2 + all_pos['qb_std_dev']**2 + all_pos['te_std_dev']**2)
```

```python
some(team_stats, 100).to_json(orient='records')
```
